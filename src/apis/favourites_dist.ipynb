```python
# Distribution of Favourites Among Art for a Certain Artist

import csv
import sys
import pandas as pd
import matplotlib.pyplot as plt

# Read CSV file into a pandas DataFrame
filename = "deviations.csv"  # replace with your actual filename
df = pd.read_csv(filename)

# Make sure there’s a 'favourites' column (you may need to adjust this)
if 'Favourites' not in df.columns:
    print("Error: 'favourites' column not found in CSV")
    sys.exit(1)

# Sort the DataFrame by 'favourites' in descending order
df_sorted = df.sort_values(by='Favourites', ascending=False)

# Plot the distribution
plt.figure(figsize=(12, 6))
plt.bar(df_sorted['Title'], df_sorted['Favourites'], color='skyblue')
plt.xticks([], [])  # removes both ticks and tick labels
# plt.xlabel('Artwork Title')
plt.ylabel('Number of Favourites')
plt.title('Distribution of Favourites Among Artworks')
plt.xticks(rotation=90, fontsize=8)
plt.tight_layout()
plt.show()
```
